#### Melihat Missing Value

```python
# memeriksa missing value
missing_value = train_data.isnull().sum()
missing_value[missing_value > 0]
```

*nb : dapat dilihat bahwa ada beberapa fitur yang memiliki missing value mendekati banyak data*

#### Mengatasi Missing Value

```python
# memisahkan data menjadi 2 bagian
# missing value > 75% dan missing value < 75%
less = missing_value[missing_value < 1000].index
over = missing_value[missing_value >= 1000].index
```

```python
# mengatasi missing value < 75% untuk data numerik menggunakan median
numeric_features = train_data[less].select_dtypes(include=['number']).columns
train_data[numeric_features] = train_data[numeric_features].fillna(train_data[numeric_features].median())
```

```python
# mengatasi missing value < 75% untuk data kategorik menggunakan modus
kategorical_features = train_data[less].select_dtypes(include=['object']).columns

for column in kategorical_features :
       train_data[column] = train_data[column].fillna(train_data[column].mode()[0])
       # menggunakan loop karena modus tiap kolom berbeda beda
```

```python
# mengatasi missing value > 75% dengan menhapus kolomnya
df = train_data.drop(columns=over)
# menyalin train data ke df agar data awal tidak mengalami perubahan karena penghapusan kolom yang memiliki missing value > 75%
```
